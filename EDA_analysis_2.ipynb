```python
df.query('Location != "Other" ')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean') \
    .plot(kind='bar', figsize=(12,5),title="Avg Coaster Speed by Location")
```
